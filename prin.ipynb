```python
from val_y_fun import *
from datetime import datetime as dt
from functools import filter
list_pac = {}
counter_eps = 0
counter_sisben = 0

while True:

    print("1.ingrese un paciente/n  2.informe de afiliacion EPS/n   3.borrar un paciente/n   4.salir/n") 
    option = input("\nSeleccione una opción: ")

    if option == "1":
        nombre=input("Ingresa el nombre dle paciente : ")

        d=validar("ingrese el dia de nacimiento : ")
        m=validar("ingrese el mes de nacimiento : ")
        a=validar("ingrese el año de nacimiento : ")
        try:
            fecha_nacimiento =dt(a,m,d)
            hoy = dt.now()
            edad = hoy.year - fecha_nacimiento.year - ((hoy.month, hoy.day) < (fecha_nacimiento.month, fecha_nacimiento.day))
            print("La edad es:", edad)

        except ValueError:
            print("ingrese la fecha correcta")
            continue

        documento=validar("Ingrese el documento del paciente: " )   
        opcion_eps = input("Seleccione la EPS del paciente:1 SISBE/n 2.Sura/n 3.Coomeva/n 4.Medimas/n 5.IPS Universitaria/n  6.Salud Total/n")    

        if opcion_eps == "1":
            eps = "SISBEN"
            counter_sisben +=1
        elif opcion_eps == "2":
            eps = "Sura"
            counter_eps += 1
        elif opcion_eps == "3":
            eps = "Coomeva"
            counter_eps += 1
        elif opcion_eps == "4":
            eps = "Medimas"
            counter_eps += 1
        elif opcion_eps == "5":
            eps = "IPS Universitaria"
            counter_eps += 1
        elif opcion_eps == "6":
            eps = "Salud Total"
            counter_eps += 1
        else:
            print("Opción no válida. Por favor, intente de nuevo.")
            continue

        
        genero = input("Ingrese el género del paciente (mujer/hombre): ")
        if genero not in ['mujer', 'hombre']:
            print("Género no válido. Por favor, ingrese 'mujer' o 'hombre'.")
            continue
        lh = validar("Ingrese el resultado de la prueba LH: ")
        diagnostico = diagnosticar_lh(genero, edad, lh)
        
        codigo = generate_patient_code(eps, counter_sisben if eps == "SISBEN" else counter_eps)
        list_pac[documento] = {nombre,codigo, fecha_nacimiento, edad, eps, (lh,diagnostico)}
        print(f"Paciente {nombre} identificado con cédula {documento} ingresado exitosamente, su código es {codigo} y su diagnostico es {diagnostico}.")
        

    elif option == "2":
        
            print("\nSubmenú de pacientes:")
            print("a) Buscar paciente")
            print("b) Ver cantidad de pacientes totales")
            print("c) Ver cantidad de pacientes menores de 10")
            print("d) Ver cantidad de pacientes mayores de 60")

            sub_option = input("\nSeleccione una opción del submenú de pacientes: ").lower()

            if sub_option == "a":
                search_patient(list_pac)

            elif sub_option == "b":
                print(f"\nCantidad total de pacientes: {len(list_pac)}")

            elif sub_option == "c":
                patients_under_10 = len(list(filter(lambda p: (dt.now() - p.fecha_nacimiento).days / 365 < 10, list_pac.values())))
                print(f"\nCantidad de pacientes menores de 10 años: {patients_under_10}")

            elif sub_option == "d":
                patients_over_60 = len(list(filter(lambda p: (dt.now() - p.fecha_nacimiento).days / 365 > 60, list_pac.values())))
                print(f"\nCantidad de pacientes mayores de 60 años: {patients_over_60}")

            else:
                break

    elif option == "3":

        cedula = input("Ingrese la cédula del paciente a borrar: ")
        if cedula in list_pac:
            del list_pac[documento]
            print("Paciente borrado exitosamente.")
        else:
            print(f"El paciente con el número de identificación {cedula} no existe en la base de datos.")

     
    elif option == "4":
        print("Saliendo del programa...")
        break

    else:
        print("Opción inválida. Por favor, seleccione una opción válida.")
```
